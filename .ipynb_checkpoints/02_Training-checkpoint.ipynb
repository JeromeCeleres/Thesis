```python
import pandas as pd
import numpy as np
import gensim
from multiprocessing import cpu_count, Pool #for multiprocessing data
cores = cpu_count()
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer 
  
lemmatizer = WordNetLemmatizer()
```

```python
df = pd.read_pickle('ejks_cleaned.pkl')
df.head()
```

```python
def parallel_df_process(df, func, n_cores=cores):
    pool = Pool(n_cores)
    df_split = np.array_split(df, n_cores)
    df_joined = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df_joined
```

```python
porter = PorterStemmer()

def stem_df(df):
    df['stemmed'] = df['cleaned_text'].apply(lambda x : [lemmatizer.lemmatize(word) for word in x])
    return df
```

```python
df_cleaned = parallel_df_process(df, stem_df)
```

```python
df_cleaned.head()
```

```python
docs = [x for x in df_cleaned['stemmed'].values]
```

```python
#Word2vec implementation
model = gensim.models.Word2Vec(docs,
                              min_count=10, 
                              workers=4,
                              size=50,
                              window=5,
                              iter = 10)
```
